```python
# Define the directory
data_directory = 'Images'
# Define the image size for preprocessing
image_size = (224, 224)
```

```python
# Load the ResNet50 model
model = ResNet50(weights='imagenet', include_top=False, input_shape=(image_size[0], image_size[1], 3))
```

```python
# Function to preprocess an image
def preprocess_image(image_path):
    img = image.load_img(image_path, target_size=image_size)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_preprocessed = preprocess_input(img_array)
    return img_preprocessed
```

```python
def extract_features(image_path):
    img_preprocessed = preprocess_image(image_path)
    features = model.predict(img_preprocessed)
    features_flattened = features.flatten()
    normalized_features = features_flattened / np.linalg.norm(features_flattened)
    return normalized_features
```

```python
def build_image_index(data_directory):
    image_index = {}
    for class_name in os.listdir(data_directory):
        class_directory = os.path.join(data_directory, class_name)
        if os.path.isdir(class_directory):
            image_index[class_name] = []
            for image_file in os.listdir(class_directory):
                image_path = os.path.join(class_directory, image_file)
                try:
                    features = extract_features(image_path)
                    image_index[class_name].append((image_path, features))
                except Exception as e:
                    print(f"Error processing image {image_path}: {str(e)}")
    return image_index
```

```python
# Function to search for similar images
def search_similar_images(query_image_path, image_index, top_n=5):
    query_features = extract_features(query_image_path)

    similar_images = []
    for class_name, images in image_index.items():
        for image_path, features in images:
            similarity_score = cosine_similarity(query_features.reshape(1, -1), features.reshape(1, -1))
            similar_images.append((image_path, similarity_score))

    similar_images.sort(key=lambda x: x[1], reverse=True)
    top_similar_images = similar_images[:5]
    return top_similar_images
```

```python
# Build the image index
image_index = build_image_index(data_directory)
```

```python
import joblib

# Save the image index to a file using joblib
index_file = 'image_index.pkl'
joblib.dump(image_index, index_file)
```
